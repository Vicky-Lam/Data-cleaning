```python
import pandas as pd
```

```python
#read excel file with multiple spreadsheets

df = pd.concat(pd.read_excel(r'dataset-demo.xlsx', sheet_name=None, engine='openpyxl'), ignore_index=True)
df
```

```python
#drop columns that do not contain useful data for analysis

df = df.drop(df.columns[[7, 8, 9, 10]], axis=1)
df
```

```python
#give new column names
df.columns = ['Data type', 'Turn taker', 'Elicit or Response', 'Code', 'Turn number', 'Speaker', 'Discourse']
df
```

```python
# fillna in column 'Discourse' with the value above
df['Discourse'].fillna(method='ffill', inplace=True)
df
```

```python
# drop rows that contain the string "Talk move"
df1 = df[~df['Code'].str.contains('Talk move', na=False)]
df1
```

```python
# drop rows if NaN appears in the column named "Code"
df2 = df1.dropna(subset=['Code'])
df2
```

```python
#drop rows with "R"
df3 = df2[~df2['Elicit or Response'].str.contains('R', na=False)]
df3
```

```python
#drop rows with "N" in the column "Elicit or Response"
df4 = df3[~df3['Elicit or Response'].str.contains('N', na=False)]
df4
```

```python
#drop rows with "P" in the column "Elicit or Response"
df5 = df4[~df4['Elicit or Response'].str.contains('P', na=False)]
df5
```

```python
# drop rows that contain "S" in the column "Turn take"
df6 = df5[~df5['Turn taker'].str.contains('S', na=False)]
df6
```

```python
#see if there is NaN in column 'Discourse'
df6['Discourse'].isnull().sum()
```

```python
#drop columns that are not useful for analysis "Turn number" and "Speaker"
df7 = df6.drop(['Turn number', 'Speaker'], axis=1)
df7
```

```python
#group the entries by the codes into the same spreadsheet and name the spreadsheet with the correponding code.

code_list = ['I1', 'D1', 'S1', 'A1', 'C1', 'R1', 'R2', 'C2', 'E1', 'C3', 'P1', 'C4', 'S2', 'R3', 'R4', 'R5', 'A2', 'C5', 'E2', 'D2', 'B1', 'P2']
with pd.ExcelWriter('cleaned-dataset-demo.xlsx') as writer:  
    for code in code_list:
        print(code)
        _temp_df = df7[df7['Code'] == code]
        _temp_df.to_excel(writer, sheet_name=code.replace("/", ""))
```
